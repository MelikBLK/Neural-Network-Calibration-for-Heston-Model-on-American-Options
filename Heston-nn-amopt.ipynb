**Geometirc Brownian Motion**

Defines the simulate method with parameters for the initial stock price (s0), time horizon (T), number of paths (n), number of discretization points (m), and initial volatility (v0, which is not used in GBM but included for compatibility). The method returns a DataFrame, although the correct return type should be a NumPy array based on the implementation.

```python
@dataclass
class GeometricBrownianMotion(StochasticProcess):

   """
    A classic geometric brownian motion which can be simulated.
    The closed form formula allow a fully vectorized calculation of the paths.
    """
mu: float #mu: The risk-free rate or the expected return of the asset.
sigma: float #kappa: The speed of reversion for the volatility to its long-term mean.

def simulate(self,s0:float,T:int,n:int,m:int,v0:float = None) -> pd.DataFrame: #n is number of paths, m is the number of discretization points.

        dt = T / m
        np.random.seed(0)
        W = np.cumsum(np.sqrt(dt) * np.random.randn(m + 1, n), axis=0)
        W[0] = 0

        T = np.ones(n).reshape(1, -1) * np.linspace(0, T, m + 1).reshape(-1, 1)

        s = s0 * np.exp((self.mu - 0.5 * self.sigma**2) * T + self.sigma * W)

        return s
```

- `s0`: Initial stock price.
- `T`: Total time horizon for the simulation.
- `n`: Number of paths to simulate (number of simulations).
- `m`: Number of discretization points (time steps).
- `v0`: Initial volatility, which is not used in the GBM model (included for compatibility with other models).
- `s`: s represents the simulated asset prices over time, following a geometric Brownian motion (GBM) model. The GBM model is a stochastic process that is widely used in finance for modeling the dynamics of various financial instruments, such as stock prices.

**Definition of Heston Process Class**


This code defines a class HestonProcess that inherits from StochasticProcess and is intended for simulating the price movements of an asset (in this case, options on Apple stock) using the Heston model. The Heston model is a stochastic volatility model used in financial mathematics to describe the evolution of an asset's price and its volatility.

- `mu`: The drift term or the risk-free rate, representing the expected return of the asset.
- `kappa`: The speed of reversion of the volatility to its mean.
- `theta`: The long-term mean volatility.
- `eta`: The volatility of the volatility, indicating how much the volatility itself can vary.
- `rho`: The correlation between the asset's returns and its volatility. This is crucial for modeling the leverage effect, where asset prices and their volatility are inversely correlated.

```python
from dataclasses import dataclass
import numpy as np
import pandas as pd
# Plus the import statement for your StochasticProcess class
```

```python
@dataclass
class HestonProcess(StochasticProcess):
    """
    An Heston process which can be simulated using Milstein schema.
    """

    mu: float
    kappa: float
    theta: float
    eta: float
    rho: float

    def simulate(
        self, s0: float, v0: float, T: int, n: int, m: int
    ) -> pd.DataFrame:  # n = number of paths, m = number of discretization points

        dt = T / m
        z1 = np.random.randn(m, n)
        z2 = self.rho * z1 + np.sqrt(1 - self.rho**2) * np.random.randn(m, n)

        s = np.zeros((m + 1, n))
        x = np.zeros((m + 1, n))
        v = np.zeros((m + 1, n))

        s[0] = s0
        v[0] = v0

        for i in range(m):

            v[i + 1] = (
                v[i]
                + self.kappa * (self.theta - v[i]) * dt
                + self.eta * np.sqrt(v[i] * dt) * z1[i]
                + self.eta**2 / 4 * (z1[i] ** 2 - 1) * dt
            )
            v = np.where(v > 0, v, -v)

            x[i + 1] = x[i] + (self.mu - v[i] / 2) * dt + np.sqrt(v[i] * dt) * z2[i]

            s[1:] = s[0] * np.exp(x[1:])

        return s
```

**Option Class Definition**

```python
@dataclass
class Option():
    s0: float
    v0: float
    T: float
    K:float
    call: bool

    def payoff(self, s: np.ndarray) -> np.ndarray:
        payoff = np.maximum(s - self.K, 0) if self.call else np.maximum(self.K - s, 0)
        return payoff
```

- `Usage`: This line calculates the payoff of the option. For a put option (self.call == False), it calculates the maximum of self.K - s and 0, representing the profit from exercising the option when the final price s is below the strike price K, and zero otherwise.


**Defining Monte Carlo Simulation Function**

```python
import numpy as np
from numpy.polynomial import Polynomial
from scipy.stats import norm


def monte_carlo_simulation(option: Option, process: StochasticProcess, n: int, m: int, alpha: float = 0.05) -> float:
    """
    Given an option and a process followed by the underlying, calculate the classic monte carlo price estimator
    """
    # n = number of paths, m = number of discretization points
    s = process.simulate(s0=option.s0, v0=option.v0, T=option.T, n=n, m=m)
    st = s[-1]
    payoffs = option.payoff(s=st)

    discount = np.exp(-process.mu * option.T)
    price = np.mean(payoffs) * discount

    quantile = norm.ppf(1 - alpha / 2)
    confidence_interval = [
        np.round(price - quantile * np.std(payoffs * discount) / np.sqrt(n), 2),
        np.round(price + quantile * np.std(payoffs * discount) / np.sqrt(n), 2),
    ]

    print(f"The price of {option!r} = {price:.2f}")
    print(f"{(1-alpha)*100}% confidence interval = {confidence_interval}")

    return np.round(price, 2)
```

**Explanation of the Monte Carlo Simulation Function**
## Monte Carlo Simulation for Option Pricing

This function estimates the price of an option using the Monte Carlo simulation method. It takes into account the dynamics of the underlying asset modeled by a given stochastic process (e.g., the Heston process).

### Parameters
- `option`: An `Option` object containing initial parameters like the initial stock price (`s0`), initial volatility (`v0`), and time to maturity (`T`).
- `process`: A `StochasticProcess` object responsible for simulating the path of the underlying asset.
- `n`: The number of paths to simulate.
- `m`: The number of discretization points per path.
- `alpha`: The significance level for the confidence interval (default is 5%).

### Procedure
1. Simulate `n` paths for the underlying asset using the specified `process`.
2. Calculate the option payoffs based on the final simulated asset prices.
3. Discount the mean payoff to present value to estimate the option price.
4. Calculate a confidence interval for the price estimate to quantify the estimation uncertainty.

The function provides a practical approach to option pricing, incorporating the randomness of the underlying asset's movements through simulation. It returns the estimated price of the option along with a confidence interval, offering insight into the precision of the estimate.


**Defining Monte Carlo Simulation x Longstaff-SchwartzFunction**
## Explanation
## Monte Carlo Simulation with Longstaff-Schwartz Algorithm

The `monte_carlo_simulation_LS` function implements the Longstaff-Schwartz algorithm for option pricing through Monte Carlo simulation. This method is particularly suited for pricing American options, where the option holder has the right to early exercise. Here's an overview of the function implementation:

### Function Inputs

- `Option` object: Contains option specifics such as initial price (`s0`), volatility (`v0`), and time to maturity (`T`).
- `StochasticProcess` object: Defines the dynamics of the underlying asset price.
- `n`: Number of Monte Carlo simulation paths.
- `m`: Number of time discretization points.
- `alpha`: Confidence level for the price estimation (default is 0.05, not used in this code block).

### Key Steps in the Algorithm

1. **Path Simulation**: Simulate `n` paths of the underlying asset's price over `m` time steps.
2. **Payoff Calculation**: For each simulated path, calculate the option's payoff at every time step.
3. **Option Value Initialization**: Initialize an array to store the option values, setting the final values to the corresponding payoffs.
4. **Discount Factor**: Compute the discount factor using the risk-free rate and the time step size.
5. **Backward Induction**: Starting from the end of the option's life, move backward in time, at each step deciding whether to exercise the option or continue holding it based on a polynomial fit of continuation values.
6. **Price Estimation**: Estimate the option price as the discounted mean of the option values at the first time step.

### Output

The function prints and returns the estimated price of the option, rounded to four decimal places.




```python
import numpy as np
from numpy.polynomial import Polynomial

def monte_carlo_simulation_LS(option: Option, process: StochasticProcess, n: int, m: int, alpha: float = 0.05) -> float:
    """
    Given an option and a process followed by the underlying, calculate the option value using the Longstaff-Schwartz algorithm.
    """
    s = process.simulate(s0=option.s0, v0=option.v0, T=option.T, n=n, m=m)

    payoffs = option.payoff(s=s)

    v = np.zeros_like(payoffs)
    v[-1] = payoffs[-1]

    dt = option.T / m
    discount = np.exp(-process.mu * dt)

    for t in range(m - 1, 0, -1):
        # Data validation and cleaning
        valid_indices = ~np.isnan(s[t]) & ~np.isinf(s[t]) & ~np.isnan(v[t + 1]) & ~np.isinf(v[t + 1])
        s_clean = s[t][valid_indices]
        v_clean = v[t + 1][valid_indices]
        discount_clean = discount * v_clean
        
        if len(s_clean) > 0 and len(v_clean) > 0:  # Ensure there's data to fit
            try:
                polynome = Polynomial.fit(s_clean, discount_clean, 5)
                c = polynome(s[t])
                v[t] = np.where(payoffs[t] > c, payoffs[t], discount * v[t + 1])
            except Exception as e:
                print(f"Error fitting polynomial at t={t}: {e}")
                # Fallback or skip this step if necessary
        else:
            print(f"No valid data for polynomial fit at t={t}. Skipping.")
            # Implement fallback logic if needed

    price = discount * np.mean(v[1])
    print(f"The price of {option!r} = {round(price, 4)}")

    return round(price, 4)
```

## Defining the Black-Scholes-Merton Function

```python
def black_scholes_merton(r, sigma, option: Option):
    """
    Calculate the price of vanilla options using BSM formula
    """
    d1 = (np.log(option.s0 / option.K) + (r + sigma**2 / 2) * option.T) / (sigma * np.sqrt(option.T))
    d2 = d1 - sigma * np.sqrt(option.T)

    price = option.s0 * norm.cdf(d1) - option.K * np.exp(-r * option.T) * norm.cdf(d2)
    price = price if option.call else price - option.s0 + option.K * np.exp(-r * option.T)

    return np.round(price, 2)
```

## Explanation
### Black-Scholes-Merton Option Pricing Function

This function utilizes the Black-Scholes-Merton (BSM) formula to calculate the price of vanilla options, both calls and puts. It's structured to handle the fundamental variables of option pricing within a financial context.

#### Parameters:
- `r`: Risk-free interest rate, indicating the return of an investment with no risk over a specific period.
- `sigma`: Volatility of the underlying asset's returns, a measure of how much the asset's price is expected to fluctuate.
- `option`: An instance of the `Option` class with attributes:
  - `s0`: Current price of the underlying asset.
  - `K`: Strike price of the option.
  - `T`: Time to expiration of the option, in the same units as `r`.
  - `call`: Boolean indicating if the option is a call (`True`) or put (`False`).

#### Formula Components:
- **`d1` and `d2`**: Intermediate calculations that incorporate the current asset price, strike price, time to expiration, risk-free rate, and volatility to derive option pricing.
- **Option Price Calculation**:
  - For a call option: `price = s0 * norm.cdf(d1) - K * exp(-r * T) * norm.cdf(d2)`
  - For a put option: Adjusts the call price to account for the option type: `price = price - s0 + K * exp(-r * T)`

The result is the option price, rounded to two decimal places.


## Defining the Cox-Ross-Rubinstein (CRR) model

```python
def crr_pricing(r=0.1, sigma=0.2, option: Option = Option(s0=100, T=1, K=100, v0=None, call=False), n=25000):
    """
    Calculate the price of an americain option using a backward tree model
    """
    dt = option.T / n
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    a = np.exp(r * dt)
    p = (a - d) / (u - d)
    q = 1 - p

    st = np.array([option.s0 * u**i * d ** (n - i) for i in range(n + 1)])
    v = np.maximum(option.K - st, 0)

    for _ in range(n):
        v[:-1] = np.exp(-r * dt) * (p * v[1:] + q * v[:-1])
        st = st * u
        v = np.maximum(v, option.K - st)

    return np.round(v[0], 3)
```

## Explanation
## American Option Pricing Using Cox-Ross-Rubinstein (CRR) Model

This section outlines the implementation of the CRR model to price American options through a binomial tree approach. The function `crr_pricing` encapsulates the logic for this calculation based on several parameters and the option's characteristics.

### Parameters

- `r`: The risk-free interest rate, annualized.
- `sigma`: The volatility of the underlying asset, annualized.
- `option`: An instance of the `Option` class, containing:
  - `s0`: Initial stock price.
  - `T`: Time to expiration in years.
  - `K`: Strike price.
  - `call`: Boolean indicating if the option is a call (False implies it's a put).
- `n`: Number of steps in the binomial tree, affecting the model's accuracy and computational complexity.

### Process Overview

1. **Initialization:**
   - Calculate the length of each time interval (`dt`) in the binomial tree.
   - Determine the up (`u`) and down (`d`) factors for stock price movement, and the risk-free growth factor (`a`).
   - Compute risk-neutral probabilities (`p` for up, `q` for down movements).

2. **Stock Price and Option Value at Maturity:**
   - Generate the possible stock prices at maturity within the binomial tree.
   - Initialize the option values at maturity based on their intrinsic value (difference between strike price and stock prices for a put option).

3. **Backward Induction:**
   - Iterate backward through the tree to calculate the option value at each step, considering the maximum value between holding and exercising the option early (for American options).

### Return Value

The function returns the calculated option price at the current time, rounded to three decimal places. This approach effectively models the early exercise feature inherent to American options, providing a flexible method to estimate their pricing.

## Applying Heston Model to Option Pricing
## Heston Model Parameters Explained

The Heston model is a sophisticated tool for option pricing that takes into account the stochastic nature of volatility. Here are the key parameters it uses:

- **S (Spot Price):** The current market price of the underlying asset, such as a stock. This value serves as the starting point for the model's simulations.

- **K (Strike Price):** The agreed-upon price at which the option holder can either buy (in the case of a call option) or sell (in the case of a put option) the underlying asset. This price is fixed and determined when the option is purchased.

- **T (Time to Maturity):** The duration remaining until the option expires, typically expressed in years. The value of an option tends to decline as it approaches its expiration date, influenced by factors such as volatility and movements in the underlying asset's price.

- **mu (Risk-free Interest Rate):** Represented by "μ", this is the theoretical return on an investment that is considered free of risk, usually based on the yield of government bonds. It's used to discount the future payoff of the option to its present value.

- **V0 (Initial Volatility):** Represents the current level of volatility in the returns of the underlying asset. In the Heston model, this initial volatility is a starting point that evolves over time following a stochastic process.

- **kappa (Mean Reversion Speed):** Determines the rate at which volatility reverts to its long-term mean value (theta). A higher value of kappa indicates a quicker reversion, suggesting that volatility shocks are short-lived.

- **theta (Long-term Variance):** The average long-term variance of the underlying asset's price. The model assumes that volatility will revert to this level over time, following the mean-reverting process.

- **eta (Volatility of Variance):** Also referred to as the "vol of vol", this parameter quantifies the volatility of the asset's volatility itself. A higher eta value implies that the underlying asset's volatility can undergo more significant changes, which in turn affects option pricing.

- **rho (Correlation):** Measures the correlation between the returns of the underlying asset and its volatility. A negative correlation is typical in equity markets and indicates that volatility tends to increase as stock prices decrease. This relationship is key in capturing asymmetric volatility patterns in the Heston model.

```python
import matplotlib.pyplot as plt
import pandas as pd
from scipy.optimize import least_squares

# Initialize a list outside the error function to track error per iteration
errors_per_iteration = []

def calibrate(init_val, market_datas):

     # Filter out rows where C_BID (PRICE) is zero
    market_datas = market_datas[market_datas["PRICE"] > 0]

    # Define bounds for the parameters to keep them within realistic ranges
    bounds = ([0.001, 0.001, 0.001, -0.999, 0.001],  # Lower bounds for kappa, theta, eta, rho, v0
          [10, 1, 2, 0.999, 1])  # Upper bounds for kappa, theta, eta, rho, v0


    def error(x):
        kappa, theta, eta, rho, v0 = x
        print('x=',kappa, theta, eta, rho, v0)
        total_error = 0.0
        for i in range(len(market_datas)):
            s0, k, market_price, r, T = market_datas.iloc[i]
            
            heston = HestonProcess(mu=r, kappa=kappa, theta=theta, eta=eta, rho=rho)
            opt = Option(s0=s0, v0=v0, T=T, K=k, call=True)
            
            # Adjusted for a balance between speed and accuracy
            heston_price = monte_carlo_simulation_LS(option=opt, process=heston, n=1000, m=252)
            individual_error = (heston_price - market_price) ** 2
            
            print(f"Option {i+1}: Market Price = {market_price}, Heston Price = {heston_price}, Error = {individual_error}")
            total_error += individual_error
        
        # Update the error list for plotting
        errors_per_iteration.append(total_error)
        return total_error

    # Optimization settings
    result = least_squares(
        error, 
        init_val, 
        bounds=bounds, 
        loss='soft_l1',  # More robust to outliers
        xtol=1e-3, 
        ftol=1e-3, 
        gtol=1e-3, 
        method='trf'  # Trust Region Reflective algorithm
    )

    # Plotting the error over iterations
    plt.figure(figsize=(10, 6))
    plt.plot(errors_per_iteration, marker='o', linestyle='-', color='b')
    plt.title('Error Over Iterations')
    plt.xlabel('Iteration')
    plt.ylabel('Cumulative Error')
    plt.grid(True)
    plt.show()

    print('Optimization result:', result)
    print('Total Errors Evolution: ', errors_per_iteration)
    print("Optimized Parameters: ",result.x)
    return result.x  # Return the optimized parameters
```
